```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = '..\Resources\\schools_complete.csv'
student_data_to_load = '..\Resources\\students_complete.csv'

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
number_schools = len(school_data_complete['school_name'].unique())

# Calculate the total number of students
number_students = len(school_data_complete['student_name'])

# Calculate the total budget (for all schools)
unique_schools = school_data_complete['school_name'].unique()
school_means = school_data_complete.groupby('school_name').mean()
total_budget = school_means['budget'].sum()

# Calculate the average math score
avg_math_score = school_data_complete['math_score'].mean()

# Calculate the average reading score
avg_reading_score = school_data_complete['reading_score'].mean()

# Calculate the overall passing rate (overall average score), 
#   i.e. (avg. math score + avg. reading score)/2
overall_passing_rate = (avg_math_score + avg_reading_score)/2

# Calculate the percentage of students with a passing math score (70 or greater)
num_passed_math = len(school_data_complete[school_data_complete['math_score'] >= 70])
percent_passed_math = num_passed_math/number_students * 100

# Calculate the percentage of students with a passing in reading score (70 or greater)
num_passed_reading = len(school_data_complete[school_data_complete['reading_score'] >= 70])
percent_passed_reading = num_passed_reading/number_students * 100

# Calculate the percentage of students with a passing  in both math and reading score (70 or greater)
num_passed_both = len(school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)])
percent_passed_both = num_passed_both/number_students * 100


# Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({'Total Schools': number_schools,
                                    'Total Students': number_students,
                                    'Total Budget': total_budget,
                                    'Average Math Score': avg_math_score,
                                    'Average Reading Score': avg_reading_score,
                                    '% Passing Math': percent_passed_math,
                                    '% Passing Reading': percent_passed_reading,
                                    '% Overall Passing Rate': percent_passed_both
                                   }, index = [0])

# Optional: give the displayed data cleaner formatting (to match example output)
def format_six_decimal_places():
    return lambda x: "{:.6f}".format(x)
def format_two_decimal_places():
    return lambda x: "{:.2f}".format(x)
    
    
district_summary_df.style.format({'Total Students': lambda x: "{:,}".format(x),
                                  'Total Budget': lambda x: "${:,.2f}".format(x),
                                  'Average Math Score': format_six_decimal_places(),
                                  'Average Reading Score': format_six_decimal_places(),
                                  '% Passing Math': format_six_decimal_places(),
                                  '% Passing Reading': format_six_decimal_places(),
                                  '% Overall Passing Rate': format_six_decimal_places()
                                 })
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Create an overview table that summarizes key metrics about each school:

# Create a unique list of the schools
school_list = school_data_complete['school_name'].unique()
school_list.sort()

overview_table = []
table_columns = ['School Type', 'Total Students', 'Total School Budget', \
                 'Per Student Budget', 'Average Math Score', 'Average Reading Score', \
                 '% Passing Math', '% Passing Reading', '% Overall Passing Rate']

# # Loop through each school and add key metrics to the overview_table list
for i in range(0, number_schools):

#     # School Name    
    school = school_list[i]
    
# Get Subset of Data
    subset_df = school_data_complete[school_data_complete['school_name'] == school]
    
# School Type
    school_type = subset_df['type'].iloc[0]
    
# Total Students
    total_students = subset_df['student_name'].count()

# Total School Budget
    total_school_budget = subset_df['budget'].iloc[0]

# Per Student Budget
    per_student_budget = total_school_budget/total_students
    
# Average Math Score
    average_math_score = subset_df['math_score'].mean()
    
# Average Reading Score
    average_reading_score = subset_df['reading_score'].mean()

# % Passing Math

#     passing_math_num = len(subset_df[subset_df["math_score"] >= 70])
    passing_math_num = subset_df[subset_df["math_score"] >= 70]['student_name'].count()
    percent_passing_math = passing_math_num/total_students * 100

# % Passing Reading
    passing_reading_num = len(subset_df[subset_df["reading_score"] >= 70])
# subset_df[subset_df["reading_score"] >= 70].count()
    percent_passing_reading =  passing_reading_num / total_students * 100
        
   
    passing_both = len(subset_df[(subset_df["math_score"] >= 70) & (subset_df["reading_score"] >= 70)])
#len(subset_df[(subset_df['math_score'] >= 70) & (subset_df['reading_score'] >= 70)])
    overall_pass_rate = passing_both/total_students *100
    
 # Append the school to the dataframe (index are the columns)
    overview_table.append({'School Type': school_type,
                    'Total Students': total_students,
                    'Total School Budget': total_school_budget, 
                    'Per Student Budget': per_student_budget,
                    'Average Math Score': average_math_score,
                    'Average Reading Score': average_reading_score,
                    '% Passing Math': percent_passing_math,
                    '% Passing Reading': percent_passing_reading,
                    '% Overall Passing Rate': overall_pass_rate})
    


# # Create a dataframe to hold the above results
overview_final = pd.DataFrame(overview_table, index = school_list, columns = table_columns)

format_dict = {'Total Students':'{:,}',
               'Total School Budget':'${0:,.2f}',
               'Per Student Budget':'${0:,.2f}'
              }

overview_final.style.format(format_dict)
#     {'Total Students': lambda x: "{:,}".format(x),
#                           "Total School Budget": lambda x: "${0:,.2f}".format(x),
#                           "Per Student Budget": "${:.2f}",
#                           'Average Math Score': "{:6f}", 
#                           'Average Reading Score': "{:6f}", 
#                           "% Passing Math": "{:6f}", 
#                           "% Passing Reading": "{:6f}"
#                         })





#overview_final
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
math_scores_by_grade_df = pd.DataFrame()

# Create list of unique grades
grades_list = school_data_complete['grade'].unique()

# Set index to school name
school_index_df = school_data_complete.set_index('school_name')

for i in range(0, len(grades_list)):
    
    # Set grade
    grade = grades_list[i]
    
    # Create a pandas series for each grade. Hint: use a conditional statement.
    reading_scores = pd.Series(school_index_df[school_index_df['grade'] == grade]['math_score'])
    
    # Group each series by school
    grouped_series = reading_scores.groupby(reading_scores.index).mean()
    
    # Combine the series into a dataframe
    math_scores_by_grade_df = pd.concat([math_scores_by_grade_df, grouped_series.rename(grade)], axis=1, sort=False)

#output

math_scores_by_grade_df = math_scores_by_grade_df[['9th', '10th', '11th', '12th']]
math_scores_by_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Copy relevant columns from overview_df
scores_by_spending = overview_df.drop(columns= ['School Type', 'Total Students', 'Total School Budget'])

# Bin the spending
scores_by_spending['Spending Ranges'] = pd.cut(scores_by_spending['Per Student Budget'], bins=spending_bins, labels=group_names)

# Drop the 'Per Student Budget' column
binned_scores_by_spending = scores_by_spending.drop(columns = ['Per Student Budget'])

# Groupby Bins column
binned_scores_by_spending = binned_scores_by_spending.groupby('Spending Ranges').mean()

# Rename index name
binned_scores_by_spending.index.names = ['Spending Ranges (Per Student)']

formatted_binned_scores_by_spending = binned_scores_by_spending.style.format({'Average Math Score': lambda x: "${:,.2f}".format(x), 
                                                                          'Average Reading Score': lambda x: "${:.2f}".format(x),
                                                                          '% Passing Math': format_two_decimal_places(),
                                                                          '% Passing Reading': format_two_decimal_places(),
                                                                          '% Overall Passing Rate': format_two_decimal_places()})

formatted_binned_scores_by_spending
```

# PyCity Schools Analysis - Conclusions

Overall view from the data below, we can conclude the following ;
1.	School spending ranges per student do not impact math and reading scores.
2.	School size has an impact on testing scores. The larger the school size, the lower the test scores and percentage of overall students passing.
3.	Charter Schools performance is stronger than those of District Schools, as they have higher testing scores and percentages of overall students passing.

However, more alaysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.
